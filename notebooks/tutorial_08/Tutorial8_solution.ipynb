# Tutorial 08 - Translation with a Sequence to Sequence Network and Attention

The goal of this tutorial is to build a neural network model to translate from French to English. 

## Compressed Teaching Overview

**Goal:** build and compare two sequence-to-sequence translators (French -> English):
1. a baseline Encoder-Decoder GRU
2. an Attention-based Decoder

**Main insights to leave with:**
- Machine translation is a conditional sequence generation task, not a single classification.
- The encoder maps an input token sequence to hidden representations; the decoder generates output autoregressively.
- Teacher forcing stabilizes and accelerates training, but creates train-test mismatch if overused.
- Attention removes the strict information bottleneck of a single context vector and improves long/complex sentence handling.
- Evaluation should include both qualitative inspection (sample translations, attention maps) and awareness of failure modes (length bias, repetition, OOV words).

### Data Processing

### Teaching note: Data pipeline intuition

This section is where model quality is won or lost.

- `normalizeString` reduces surface variability (case, punctuation spacing, non-letter cleanup), making token statistics more learnable.
- `filterPair` constrains sentence length and pattern complexity for a first working system. This intentionally simplifies the task so training converges quickly.
- `Lang` builds a compact vocabulary mapping; special tokens `SOS`, `EOS`, and `PAD` define decoding start, stopping criterion, and batching behavior.
- The custom `collate_fn` + `pad_sequence` lets us batch variable-length sequences efficiently.

```python
SOS_token = 0
EOS_token = 1
PAD_token = 2

class Lang:
    def __init__(self, name):
        self.name = name
        self.word2index = {}
        self.word2count = {}
        self.index2word = {0: "SOS", 1: "EOS", 2: "PAD"}
        self.n_words = 3  # Count SOS and EOS

    def addSentence(self, sentence):
        for word in sentence.split(' '):
            self.addWord(word)

    def addWord(self, word):
        if word not in self.word2index:
            self.word2index[word] = self.n_words
            self.word2count[word] = 1
            self.index2word[self.n_words] = word
            self.n_words += 1
        else:
            self.word2count[word] += 1
```

```python
# Turn a Unicode string to plain ASCII, thanks to
# https://stackoverflow.com/a/518232/2809427
def unicodeToAscii(s):
    return ''.join(
        c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) != 'Mn'
    )
# Lowercase, trim, and remove non-letter characters
def normalizeString(s):
    s = unicodeToAscii(s.lower().strip())
    s = re.sub(r"([.!?])", r" \1", s)
    s = re.sub(r"[^a-zA-Z.!?]+", r" ", s)
    return s
```

```python
def readLangs(lang1, lang2, reverse=False):
    print("Reading lines...")

    # Read the file and split into lines
    lines = open('%s-%s.txt' % (lang1, lang2), encoding='utf-8').\
        read().strip().split('\n')

    # Split every line into pairs and normalize
    pairs = [[normalizeString(s) for s in l.split('\t')] for l in lines]

    # Reverse pairs, make Lang instances
    if reverse:
        pairs = [list(reversed(p)) for p in pairs]
        input_lang = Lang(lang2)
        output_lang = Lang(lang1)
    else:
        input_lang = Lang(lang1)
        output_lang = Lang(lang2)

    return input_lang, output_lang, pairs
```

```python
MAX_LENGTH = 10
eng_prefixes = (
    "i am ", "i m ",
    "he is", "he s ",
    "she is", "she s ",
    "you are", "you re ",
    "we are", "we re ",
    "they are", "they re "
)
def filterPair(p):
    return len(p[0].split(' ')) < MAX_LENGTH and \
        len(p[1].split(' ')) < MAX_LENGTH and \
        p[1].startswith(eng_prefixes)

def filterPairs(pairs):
    return [pair for pair in pairs if filterPair(pair)]
```

```python
def prepareData(lang1, lang2, reverse=False):
    input_lang, output_lang, pairs = readLangs(lang1, lang2, reverse)
    print("Read %s sentence pairs" % len(pairs))
    pairs = filterPairs(pairs)
    print("Trimmed to %s sentence pairs" % len(pairs))
    print("Counting words...")
    for pair in pairs:
        input_lang.addSentence(pair[0])
        output_lang.addSentence(pair[1])
    print("Counted words:")
    print(input_lang.name, input_lang.n_words)
    print(output_lang.name, output_lang.n_words)
    return input_lang, output_lang, pairs


input_lang, output_lang, pairs = prepareData('eng', 'fra', True)
print(random.choice(pairs))
```

```python
def indexesFromSentence(lang, sentence):
    return [lang.word2index[word] for word in sentence.split(' ')]

def tensorFromSentence(lang, sentence):
    indexes = indexesFromSentence(lang, sentence)
    indexes.append(EOS_token)
    return torch.tensor(indexes, dtype=torch.long, device=device).view(-1, 1)

def tensorsFromPair(pair):
    input_tensor = tensorFromSentence(input_lang, pair[0])
    target_tensor = tensorFromSentence(output_lang, pair[1])
    return (input_tensor, target_tensor)
```

```python
class LangPairsDataset(Dataset):
    def __init__(self, pairs: List):
        super(LangPairsDataset, self).__init__()

        self.pairs = pairs

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, index: int):
        return tensorsFromPair(self.pairs[index])
```

```python
lang_pairs_dataset = LangPairsDataset(pairs)
print(lang_pairs_dataset[20])
print(lang_pairs_dataset.pairs[20])
```

```python
def get_dataloader(dataset, num_iters, batch_size):
    sampler = torch.utils.data.RandomSampler(
        dataset,
        replacement=True,
        num_samples=num_iters * batch_size,
        generator=torch.Generator().manual_seed(42))
    return DataLoader(dataset=dataset,
                      batch_size=batch_size,
                      collate_fn=PairsCollate(),
                      sampler=sampler)


class PairsCollate(object):
    def __init__(self):
        super(PairsCollate, self).__init__()

    @staticmethod
    def __call__(items):
        items1 = [s[0] for s in items]
        items2 = [s[1] for s in items]
        batch1 = pad_sequence(items1, batch_first=True, padding_value=PAD_token)
        batch2 = pad_sequence(items2, batch_first=True, padding_value=PAD_token)

        return batch1.squeeze(-1), batch2.squeeze(-1)
```

## Model

### Teaching note: Baseline vs attention architecture

Key conceptual comparison:

- **Baseline decoder (`DecoderRNN`)** only receives recurrent hidden state history. In classic Seq2Seq, this can force too much source information into a compressed representation.
- **Attention decoder (`AttnDecoderRNN`)** computes query-key similarity over all encoder time steps and builds a context vector per decoding step.
- This converts "one fixed context" into "dynamic retrieval", which usually helps with longer sentences and word reordering.

Shape story to highlight:
- Encoder outputs: `[b, src_len, h]`
- Decoder step/query: `[b, tgt_len, h]`
- Attention weights: `[b, tgt_len, src_len]`
- Context after weighted sum: `[b, tgt_len, h]`

```python
class EncoderRNN(nn.Module):
    def __init__(self, vocab_size, hidden_size):
        super(EncoderRNN, self).__init__()
        self.hidden_size = hidden_size

        #Embedding layer
        self.embedding = nn.Embedding(vocab_size, hidden_size)
        self.gru = nn.GRU(hidden_size, hidden_size, batch_first=True)

    def forward(self, input, hidden=None):
        """
        :param input: [b l]
        :param hidden: [n b h]
        :return: [b l h], [1 b h]
        """
        
        #Convert word index to word embedding
        embedded = self.embedding(input)  # [b l h]
        
        #forward through the GRU cell
        output, hidden = self.gru(embedded, hidden)
        
        return output, hidden
```

```python
class DecoderRNN(nn.Module):
    def __init__(self, hidden_size, output_size):
        super(DecoderRNN, self).__init__()
        self.hidden_size = hidden_size
        
        #Embedding layer
        self.embedding = nn.Embedding(output_size, hidden_size)
        #RNN layer
        self.gru = nn.GRU(hidden_size, hidden_size, batch_first=True)
        #Projection layer
        self.out = nn.Linear(hidden_size, output_size)
    
    def forward(self, input, encoder_outputs, hidden=None):
        """
        :param input: [b l]
        :param encoder_outputs: [b k h]
        :param hidden: [n b h]
        :return: [b l v], [1 b h]
        """
        
        #Convert word index to word embedding
        output = self.embedding(input) 
        output = F.relu(output)
        
        #Apply the GRU
        output, hidden = self.gru(output, hidden)
        
        #Get the prediction probabilities over target language tokens
        output = self.out(output)
        
        return output, hidden, None
```

```python
class AttnDecoderRNN(nn.Module):
    def __init__(self, hidden_size, output_size):
        super(AttnDecoderRNN, self).__init__()
        self.hidden_size = hidden_size
        self.output_size = output_size

        #Embedding layer
        self.embedding = nn.Embedding(self.output_size, self.hidden_size)
        
        #Attention layer to learn the attention weights
        self.q = nn.Linear(2 * self.hidden_size, self.hidden_size)
        self.k = nn.Linear(self.hidden_size, self.hidden_size)
        self.v = nn.Linear(self.hidden_size, self.hidden_size)
        
        # Layer to combine the input and the context vector learned through attention
        self.attn_combine = nn.Linear(self.hidden_size * 2, self.hidden_size)
        
        #GRU layer
        self.gru = nn.GRU(self.hidden_size, self.hidden_size, batch_first=True)
        
        #Projection layer
        self.out = nn.Linear(self.hidden_size, self.output_size)

    def forward(self, input, encoder_outputs, hidden=None):
        """
        :param input: [b l]
        :param encoder_outputs: [b k h]
        :param hidden: [n b h]
        :return: [b l v], [1 b h], [b l k]
        """
        
        #Convert word index to word embedding
        embedded = self.embedding(input)  # [b l h]     
    
        #Learn attention weights
        q = self.q(torch.cat([embedded, rearrange(hidden, "n b h -> b 1 (n h)").expand(-1, embedded.size(1), -1)], dim=2))
        k = self.k(encoder_outputs)
        v = self.v(encoder_outputs)
        attn_weights = F.softmax(torch.bmm(q, k.permute(0, 2, 1)), dim=2)      
        
        # Get the context vector as a weighted combination of the encoder outputs using attention weights
        attn_applied = torch.bmm(attn_weights, v)  # [b l h]     
        
        output = torch.cat([embedded, attn_applied], 2)
        
        output = self.attn_combine(output)
        output = F.relu(output)
        
        #Apply the GRU
        output, hidden = self.gru(output, hidden)
        
        #Get the prediction probabilities over target language tokens
        output = self.out(output)
        
        return output, hidden, attn_weights
```

## Training

### Teaching note: Training mechanics and stability

What happens each batch:
1. Encode source sequence.
2. Start decoder with `SOS`.
3. Unroll decoder token-by-token.
4. Compute cross-entropy at each step and average over target length.
5. Backpropagate through full sequence graph.

Important training ideas:
- **Teacher forcing ratio** controls whether next decoder input is gold token or model prediction.
- Higher teacher forcing speeds early learning; lower forcing can improve robustness at inference.
- SGD with many iterations here is intentionally simple for educational transparency.

Watch-outs to discuss:
- Exposure bias from heavy teacher forcing.
- Small batch size can produce noisy learning curves.
- Loss reduction does not always imply fluent translations.

```python
def train_batch(source, target, encoder, decoder, encoder_optimizer, decoder_optimizer, criterion, teacher_forcing_ratio=0.5):
    """    
    :param source: [b l]
    :param target: [b k]
    """

    #Zero the gradients
    encoder_optimizer.zero_grad()
    decoder_optimizer.zero_grad()
        
    ## Encode the input Sequence and store the intermediate encoder outputs
    encoder_outputs, encoder_hidden = encoder(source)  # [b l h], [1 l h]
        
    ## Initialize the first decoder input to SOS token
    sos_tokens = torch.tensor([SOS_token], device=device).unsqueeze(0).expand(target.size(0), -1)
    
    use_teacher_forcing = True if random.random() < teacher_forcing_ratio else False

    loss = 0
    decoder_input = sos_tokens
    decoder_hidden = encoder_hidden
    for di in range(target.size(1)):
        decoder_output, decoder_hidden, _ = decoder(decoder_input, encoder_outputs, decoder_hidden)  # [b 1 h], [1 b h]
        
        loss += criterion(decoder_output.squeeze(1), target[:, di])
        
        decoder_input = target[:, [di]] if use_teacher_forcing else torch.argmax(decoder_output, dim=2).detach()
        
    loss /= target.size(1)

    # Backpropagate the loss
    loss.backward()

    # Optimizers step
    encoder_optimizer.step()
    decoder_optimizer.step()

    return loss.item()
```

```python
def train(encoder, decoder, pairs, n_iters, batch_size, plot_every=100, learning_rate=0.01):
    encoder.train()
    decoder.train()
    
    plot_losses = []
    plot_loss_total = 0  # Reset every plot_every
    
    # Define optimizers
    encoder_optimizer = optim.SGD(encoder.parameters(), lr=learning_rate)
    decoder_optimizer = optim.SGD(decoder.parameters(), lr=learning_rate)

    # Define dataloaders
    dataset = LangPairsDataset(pairs)
    dataloader = get_dataloader(dataset, n_iters, batch_size)
    
    # Define the loss
    criterion = nn.CrossEntropyLoss()

    for i, (source, target) in enumerate(dataloader):
        #train and compute loss
        loss = train_batch(source, target, encoder, decoder, encoder_optimizer, decoder_optimizer, criterion)
        
        plot_loss_total += loss
        
        if (i + 1) % plot_every == 0:
            plot_loss_avg = plot_loss_total / plot_every
            plot_losses.append(plot_loss_avg)
            plot_loss_total = 0

            clear_output(wait=True)
            x = np.arange(0, len(plot_losses) * plot_every, plot_every)
            plt.plot(x, plot_losses)
            plt.show()
```

```python
%matplotlib inline

hidden_size = 256
encoder = EncoderRNN(input_lang.n_words, hidden_size).to(device)
decoder = DecoderRNN(hidden_size, output_lang.n_words).to(device)
train(encoder, decoder, pairs, 30000, 2, learning_rate=0.05, plot_every=100)
```

```python
%matplotlib inline

hidden_size = 256
attn_encoder = EncoderRNN(input_lang.n_words, hidden_size).to(device)
attn_decoder = AttnDecoderRNN(hidden_size, output_lang.n_words).to(device)
train(attn_encoder, attn_decoder, pairs, 30000, 2, learning_rate=0.05, plot_every=100)
```

## Evaluation

```python
def evaluate(encoder, decoder, sentence, max_length=MAX_LENGTH):
    encoder.eval()
    decoder.eval()
    
    #torch eval mode
    with torch.no_grad():
        #Encode the sentence to sequence of ints
        input_tensor = sentence.unsqueeze(0)
        source = input_tensor.to(device)
        
        ## Encode the input Sequence and store the intermediate encoder outputs
        encoder_outputs, encoder_hidden = encoder(source)  # [b l h], [1 l h]
            
        ## Initialize the first decoder input to SOS token
        sos_tokens = torch.tensor([SOS_token], device=device).unsqueeze(0)
        
        decoder_input = sos_tokens
        decoder_hidden = encoder_hidden
        decoded_words = []
        decoder_attentions = torch.zeros(max_length, source.size(1), device=device)
        ## Forward of the output Sequence and store the translations
        for di in range(max_length):
            decoder_output, decoder_hidden, decoder_attention = decoder(
                decoder_input, encoder_outputs, decoder_hidden)
            if decoder_attention is not None:
                decoder_attentions[di] = decoder_attention.data
            
            predicted_word = torch.argmax(decoder_output, dim=2)
            decoded_words.append(predicted_word.item())
            if decoded_words[-1] == EOS_token:
                break

            # Use the current prediction as input for the next prediction
            decoder_input = predicted_word

        return decoded_words, decoder_attentions[:di + 1]
```

```python
def evaluateRandomly(encoder, decoder, n=3):
    for i in range(n):
        pair = random.choice(pairs)
        print('>', pair[0])
        print('=', pair[1])
        t, _ = tensorsFromPair(pair)
        output_words, attentions = evaluate(encoder, decoder, t.squeeze(1), max_length=20)
        output_words = [output_lang.index2word[w] for w in output_words if w != EOS_token and w != PAD_token]
        output_sentence = ' '.join(output_words)
        print('<', output_sentence)
        print('')
```

```python
def showAttention(input_sentence, output_words, attentions):
    # Set up figure with colorbar
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(attentions.numpy(), cmap='bone')
    fig.colorbar(cax)

    # Set up axes
    ax.set_xticklabels([''] + input_sentence.split(' ') +
                       ['<EOS>'], rotation=90)
    ax.set_yticklabels([''] +  output_words)

    # Show label at every tick
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1))

    plt.show()

def evaluateAndShowAttention(input_sentence):
    t, _ = tensorsFromPair((input_sentence, "i m"))
    output_words, attentions = evaluate(
        attn_encoder, attn_decoder, t.squeeze(1))
    output_words = [output_lang.index2word[w] for w in output_words if w != EOS_token and w != PAD_token]
    print('input =', input_sentence)
    print('output =', ' '.join(output_words))
    showAttention(input_sentence, output_words, attentions.cpu())
```

## End-of-Tutorial Takeaways

- Seq2Seq translation is autoregressive generation conditioned on an encoded source sequence.
- Attention provides token-level access to encoder states and usually outperforms a plain decoder on harder sentences.
- Preprocessing decisions, batching strategy, and decoding setup are as important as model definition.
- Qualitative outputs + attention maps are excellent for learning, but production evaluation needs aggregate metrics and robust test sets.

If students remember one sentence: **attention turns translation from compress-then-generate into retrieve-while-generating**.
